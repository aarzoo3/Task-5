```python
import seaborn as sns
```

```python
labels = ['TEAM1', 'TEAM2']
sns.set()
plt.figure(figsize = (20,9))

plt.plot(data['team1'].value_counts(),color ='midnightblue')
plt.plot(data['team2'].value_counts(),color= 'crimson')

plt.title('Number of matches by each team',fontsize =17,fontweight = 'bold' )

plt.ylabel('No. of match')
plt.xlabel('Teams')

plt.xticks(rotation=90)
plt.legend(labels = labels,fontsize = 'large')


plt.show()
```
